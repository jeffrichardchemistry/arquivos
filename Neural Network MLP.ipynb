```python
import pandas as pd
import numpy as np 
from pyADA import Smetrics
```

```python
path = '/dados/ShareGDrive/MachineLearningRateConstant/dados/kOH_RC_FP.csv'
df = pd.read_csv(path)
df
```

```python
sm = Smetrics()
rc_norm = sm.rescale_norm(df['kOH'])
rc_norm
```

```python
df.insert(2, 'kOH_norm', rc_norm)
df
```

```python
xi = df.iloc[:, 4:].values.astype(int)
y = df.iloc[:, 2].values
```

# Criando Redes neurais com o tensorflow

```python
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Dense
```

```python
xi.shape[1]
```

```python
def create_net():
    net = Sequential()
    net.add(Dense(units=128, activation='relu', input_dim=1024))
    net.add(Dense(units=64, activation='relu'))
    net.add(Dense(units=1, activation='linear')) #regressÃ£o deixamos a fun.ativaÃ§Ã£o linear, para nao fazer nada
    
    net.compile(loss = 'mean_squared_error', optimizer = 'adam',
                metrics = ['mean_squared_error'])
    
    return net
```

```python
net = create_net()
```

```python
history = net.fit(xi, y,
                  batch_size=8,
                  epochs=100)
```

```python
import seaborn as sns
```

```python
loss = history.history['loss']
sns.lineplot(x=np.arange(0, len(loss), 1), y=loss)
```
